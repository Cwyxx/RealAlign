```python
#### Convert csv_format (uid, prompt, win_image_path) into uid(dir), images.png, caption.txt ####
import os
import shutil
import pandas as pd

base_target_dir = "/data_center/data2/dataset/chenwy/21164-data/dpo_dataset/pick-a-pic-v2/DiffusionDRO-HPDv3-test"
csv_file_path = "/data_center/data2/dataset/chenwy/21164-data/dpo_dataset/u2net_next_inpainting/HPDv3/top_512_images_no_anime_colorfulness_pickscore_0.02-hpdv3_all-uids.csv"
df = pd.read_csv(csv_file_path).head(12)

for index, row in df.iterrows():
    uid = row["uid"]
    prompt = row["prompt"]
    win_image_path = row["win_image_path"]
    target_dir = os.path.join(base_target_dir, uid)
    os.makedirs(target_dir, exist_ok=True)
    image_path = os.path.join(target_dir, "images.png")
    caption_path = os.path.join(target_dir, "caption.txt")
    shutil.copy(win_image_path, image_path)
    with open(caption_path, "w") as f:
        f.write(prompt)
```

```python
#### Build {uid, prompt, win_image_path, lose_image_path} from {uid, prompt} file ####
import os
import pandas as pd

base_real_image_dir = "/data_center/data2/dataset/chenwy/21164-data/dpo_dataset/u2net_next_inpainting/pick-a-pic-v2-top_500_images/real"
base_fake_image_dir = "/data_center/data2/dataset/chenwy/21164-data/dpo_dataset/u2net_next_inpainting/pick-a-pic-v2-top_500_images/fake"
source_csv_file_path = "/data_center/data2/dataset/chenwy/21164-data/dpo_dataset/u2net_next_inpainting/pick-a-pic-v2-top_500_images/top_500_images-backup.csv"
target_csv_file_path = "/data_center/data2/dataset/chenwy/21164-data/dpo_dataset/u2net_next_inpainting/pick-a-pic-v2-top_500_images/top_500_images.csv"
df = pd.read_csv(source_csv_file_path, dtype={"uid": str})

df["win_image_path"] = None
df["lose_image_path"] = None

for index, row in df.iterrows():
    uid = row["uid"]
    prompt = row["prompt"]
    
    real_image_path = os.path.join(base_real_image_dir, f"{uid}.png")
    fake_image_path = os.path.join(base_fake_image_dir, f"{uid}.png")
    
    if os.path.exists(real_image_path) and os.path.exists(fake_image_path):
        df.loc[index, "win_image_path"] = real_image_path
        df.loc[index, "lose_image_path"] = fake_image_path

df.to_csv(target_csv_file_path, index=False)
print(f"已保存结果到: {target_csv_file_path}")
```
