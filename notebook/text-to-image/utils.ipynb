```python
#### updtae the lose_image_path in csv file ####
import os
import pandas as pd

source_csv_path = "/data_center/data2/dataset/chenwy/21164-data/dpo_dataset/u2net_next_inpainting/HPDv3/top_512_images_no_anime_colorfulness_pickscore_0.02-hpdv3_all-uids.csv"
target_csv_path = "/data_center/data2/dataset/chenwy/21164-data/dpo_dataset/text-to-image/HPDv3/top_512_images_no_anime_colorfulness_pickscore_0.02-hpdv3_all-uids.csv"
base_fake_image_dir = "/data_center/data2/dataset/chenwy/21164-data/dpo_dataset/text-to-image/HPDv3/fake"

df = pd.read_csv(source_csv_path)

for index, row in df.iterrows():
    uid = row["uid"]
    fake_image_path = os.path.join(base_fake_image_dir, f"{uid}.png")
    if os.path.exists(fake_image_path):
        df.loc[index, "lose_image_path"] = fake_image_path
    else:
        print(f"Fake image not found for uid: {uid}")

df.to_csv(target_csv_path, index=False)
```
