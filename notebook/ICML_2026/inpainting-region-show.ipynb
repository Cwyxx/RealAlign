```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

uid = "5d41c713-22fc-44da-8b81-d528254cad76"
base_real_dir = "/data_center/data2/dataset/chenwy/21164-data/dpo_dataset/u2net_next_inpainting/HPDv3/real"
base_fake_dir = "/data_center/data2/dataset/chenwy/21164-data/dpo_dataset/u2net_next_inpainting/HPDv3/fake"
base_mask_dir = "/data_center/data2/dataset/chenwy/21164-data/dpo_dataset/u2net_next_inpainting/HPDv3/mask"

real_image_path = os.path.join(base_real_dir, f"{uid}.png")
fake_image_path = os.path.join(base_fake_dir, f"{uid}.png")
mask_image_path = os.path.join(base_mask_dir, f"{uid}.png")

# 加载图片
real_image = cv2.imread(real_image_path)
real_image = cv2.cvtColor(real_image, cv2.COLOR_BGR2RGB)

fake_image = cv2.imread(fake_image_path)
fake_image = cv2.cvtColor(fake_image, cv2.COLOR_BGR2RGB)

mask_image = cv2.imread(mask_image_path, cv2.IMREAD_GRAYSCALE)

# 提取mask的边缘轮廓
# 使用Canny边缘检测或者findContours
contours, _ = cv2.findContours(mask_image, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# 创建带边缘轮廓的real image
real_with_contour = real_image.copy()
cv2.drawContours(real_with_contour, contours, -1, (255, 0, 0), 4)  # 蓝色轮廓，线宽2

# 创建带边缘轮廓的fake image
fake_with_contour = fake_image.copy()
cv2.drawContours(fake_with_contour, contours, -1, (255, 0, 0), 4)  # 蓝色轮廓，线宽2

# 显示结果
fig, axes = plt.subplots(1, 2, figsize=(15, 7))
axes[0].imshow(real_with_contour)
axes[0].set_title('Real Image with Mask Contour')
axes[0].axis('off')

axes[1].imshow(fake_with_contour)
axes[1].set_title('Fake Image with Mask Contour')
axes[1].axis('off')

plt.tight_layout()
plt.show()

# 保存图片
# 将RGB转换回BGR格式以便保存
real_with_contour_bgr = cv2.cvtColor(real_with_contour, cv2.COLOR_RGB2BGR)
fake_with_contour_bgr = cv2.cvtColor(fake_with_contour, cv2.COLOR_RGB2BGR)

real_save_path = os.path.join(f"real-{uid}-with_contour.png")
fake_save_path = os.path.join(f"fake-{uid}-with_contour.png")

cv2.imwrite(real_save_path, real_with_contour_bgr)
cv2.imwrite(fake_save_path, fake_with_contour_bgr)

print(f"Real image with contour saved to: {real_save_path}")
print(f"Fake image with contour saved to: {fake_save_path}")
```
