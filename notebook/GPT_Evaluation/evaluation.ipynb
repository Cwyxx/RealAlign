```python
import os
import pandas as pd
base_result_dir = "/data_center/data2/dataset/chenwy/21164-data/diffusion-dpo/sd-3-5-medium/GPT_Evaluation_v3/HPDv2-photo/civitai-vs-DiffusionNFT"
MLLMs = ["gpt-5.1-chat-latest", "gemini-3-flash-preview", "claude-haiku-4-5-20251001"]
MLLMs_dict = {}
for MLLM in MLLMs:
    csv_path = os.path.join(base_result_dir, f"{MLLM}_comparison_results.csv")
    MLLMs_dict[MLLM] = pd.read_csv(csv_path)
    
matched_prompt_list = []
prompt_information_list = []
gpt_5_1_chat_latest_df = MLLMs_dict["gpt-5.1-chat-latest"]
for index, row in gpt_5_1_chat_latest_df.iterrows():
    prompt = row["prompt"]
    prompt_information = {
        "prompt": prompt,
        "gpt-5.1-chat-latest": None,
        "gemini-3-flash-preview": None,
        "claude-haiku-4-5-20251001": None
    }
    for MLLM in MLLMs:
        df = MLLMs_dict[MLLM]
        matching_rows = df[df["prompt"] == prompt]
        
        if not matching_rows.empty:
            try:
                selection_value = int(matching_rows.iloc[0]["selection"])
                prompt_information[MLLM] = selection_value
            except:
                print(f"{MLLM}: Invalid selection value: {matching_rows.iloc[0]['selection']}")
    prompt_information_list.append(prompt_information)

win_prompt_list = []
valid_prompt_list = []
for prompt_information in prompt_information_list:
    is_valid_prompt = True
    for MLLM in MLLMs:
        if prompt_information[MLLM] is None:
            is_valid_prompt = False
            break
    if is_valid_prompt:
        valid_prompt_list.append(prompt_information)
        selection_1 = sum([ prompt_information[MLLM] == 1 for MLLM in MLLMs ])
        selection_2 = sum([ prompt_information[MLLM] == 2 for MLLM in MLLMs ])
        if selection_1 > selection_2:
            win_prompt_list.append(prompt_information)
win_rate = len(win_prompt_list) / len(valid_prompt_list)
print(f"win: {len(win_prompt_list)}, total: {len(valid_prompt_list)}")
print(f"win rate: {win_rate}")
```
