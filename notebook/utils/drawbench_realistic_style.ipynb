```python
import os
source_drawbench_prompt_path = "/data3/chenweiyan/notebook/fine-tune-diffusion/spo_gitee/DiffusionNFT/dataset/drawbench/test.txt"
target_drawbench_prompt_path = "/data3/chenweiyan/notebook/fine-tune-diffusion/spo_gitee/DiffusionNFT/dataset/drawbench_realistic_style/test.txt"
os.makedirs(os.path.dirname(target_drawbench_prompt_path), exist_ok=True)
with open(source_drawbench_prompt_path, 'r', encoding='utf-8') as f:
    lines = f.readlines()

modified_lines = []
for line in lines:
    line = line.strip()  # 去除行尾的换行符和空白字符
    if line:  # 如果行不为空
        modified_line = f"Realistic photo {line}\n"
        modified_lines.append(modified_line)

# 保存到目标文件
with open(target_drawbench_prompt_path, 'w', encoding='utf-8') as f:
    f.writelines(modified_lines)

print(f"处理完成！共处理 {len(modified_lines)} 行")
print(f"已保存到: {target_drawbench_prompt_path}")
```
