```python
import os
import random
random.seed(42)

source_image_dir="/data_center/data2/dataset/chenwy/21164-data/coco_2014/val2014-10k"
target_image_dir="/data3/chenweiyan/notebook/fine-tune-diffusion/spo_gitee/tmp/SD_3_5_medium/real"
os.makedirs(target_image_dir, exist_ok=True)

image_list = os.listdir(source_image_dir)
sampled_images = random.sample(image_list, 1000)
for image in sampled_images:
    source_path = os.path.join(source_image_dir, image)
    target_path = os.path.join(target_image_dir, image)
    os.symlink(source_path, target_path)
```
