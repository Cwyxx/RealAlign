```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt

image_type = "fake"
image_id = "/data_center/data2/dataset/chenwy/21164-data/dpo_dataset/u2net_next_inpainting/civitai-top-sfw-images-with-metadata/visualization/top_512_images_no_anime_colorfulness_pickscore_0.02-hpdv3_all-uids/real/362.png"
image_id = os.path.basename(image_id).split(".")[0]
image_path = f"/data_center/data2/dataset/chenwy/21164-data/dpo_dataset/u2net_next_inpainting/civitai-top-sfw-images-with-metadata/{image_type}/{image_id}.png"
mask_path = f"/data_center/data2/dataset/chenwy/21164-data/dpo_dataset/u2net_next_inpainting/civitai-top-sfw-images-with-metadata/mask/{image_id}.png"

# image_path = f"/data_center/data2/dataset/chenwy/21164-data/dpo_dataset/u2net_next_inpainting/HPDv3/{image_type}/{image_id}.png"
# mask_path = f"/data_center/data2/dataset/chenwy/21164-data/dpo_dataset/u2net_next_inpainting/HPDv3/mask/{image_id}.png"


# 检查mask文件是否存在
if not os.path.exists(mask_path):
    print(f"Warning: Mask file not found at {mask_path}")
else:
    # 加载图像
    image = cv2.imread(image_path)
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    
    # 加载mask（灰度图）
    mask_image = cv2.imread(mask_path, cv2.IMREAD_GRAYSCALE)
    
    if mask_image is None:
        print(f"Error: Failed to load mask from {mask_path}")
    else:
        # 提取mask的边缘轮廓
        contours, _ = cv2.findContours(mask_image, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
        
        # 创建带红色轮廓的图像
        image_with_contour = image.copy()
        # 红色在RGB格式中是 (255, 0, 0)，线宽设为3
        cv2.drawContours(image_with_contour, contours, -1, (255, 0, 0), 3)
        
        # 显示结果
        plt.figure(figsize=(10, 10), dpi=300)
        plt.imshow(image_with_contour)
        plt.axis('off')
        plt.tight_layout()
        plt.show()
        
        print(f"Successfully outlined salient regions.")

print(f"image_id: {image_id}")
```
