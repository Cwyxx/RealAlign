```python
import os
import shutil

base_image_dir = "/data_center/data2/dataset/chenwy/21164-data/diffusion-dpo/sd-3-5-medium/generate_images"
base_target_image_dir = "/data3/chenweiyan/notebook/fine-tune-diffusion/spo_gitee/tmp/user_study-SD-3.5-M"

# dataset_name = "HPDv2-anime"
# id_list = [0, 2, 5, 9, 43, 61, 72, 87, 117, 149, 150, 156, 172, 193, 198]

# dataset_name = "HPDv2-concept-art"
# id_list = [14, 18, 28, 60, 69, 78, 82, 105, 152, 174, 181, 191, 1, 9, 91]

# dataset_name = "HPDv2-paintings"
# id_list = [3, 7, 9, 16, 24, 25, 27, 36, 37, 56, 93, 94, 120, 157, 163]

dataset_name = "HPDv2-photo"
id_list = [19, 26, 47, 58, 66, 67, 70, 71, 84, 87, 100, 110, 146, 148, 152]

method_list = [
    "sd-3-5-medium",
    "pick-a-pic-v2-dpo_dataset_160000_pairs",
    "ours"
]
method_dict = {
    "sd-3-5-medium": "sd-3-5-medium/ckpt-0/images",
    "pick-a-pic-v2-dpo_dataset_160000_pairs": "pick-a-pic-v2-dpo_dataset_160000_pairs/ckpt-4900/images",
    "ours": "irl_top_512_images_no_anime_colorfulness_pickscore_0.02-hpdv3_all_lr_0.0002_ckpt_3200-dpo_top_512_images_no_anime_colorfulness_pickscore_0.02-hpdv3_all/ckpt-450/images",
}

os.makedirs(base_target_image_dir, exist_ok=True)

EXTS = [".png", ".jpg", ".jpeg", ".webp"]

for method in method_list:
    src_dir = os.path.join(base_image_dir, dataset_name, method_dict[method])

    target_dir = os.path.join(base_target_image_dir, dataset_name, method)
    os.makedirs(target_dir, exist_ok=True)

    print(f"Processing method: {method}")
    print(f"Source dir: {src_dir}")

    for img_id in id_list:
        found = False

        for ext in EXTS:
            filename = f"{img_id:05d}{ext}"
            src_path = os.path.join(src_dir, filename)

            if os.path.exists(src_path):
                dst_path = os.path.join(target_dir, filename)

                shutil.copy2(src_path, dst_path)
                print(f"Copied: {src_path} -> {dst_path}")

                found = True
                break

        if not found:
            print(f"[WARNING] Image {img_id} not found in {src_dir}")
```
