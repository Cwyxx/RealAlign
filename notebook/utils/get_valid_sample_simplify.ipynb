```python
#### Get highest real image pickscore and anime is no and pickscore (gap > 0.02) uid ####
import os
import pandas as pd
import matplotlib.pyplot as plt

all_information_csv_file_path = "/data_center/data2/dataset/chenwy/21164-data/dpo_dataset/HPDv3/real_images_uid_prompt.csv"
anime_csv_file_path = "/data_center/data2/dataset/chenwy/21164-data/dpo_dataset/u2net_next_inpainting/HPDv3/qwen3_anime/qwen3_anime.csv"
colorfulness_csv_file_path = "/data_center/data2/dataset/chenwy/21164-data/dpo_dataset/u2net_next_inpainting/HPDv3/colorfulness/colorfulness_score.csv"
pickscore_csv_file_path = "/data_center/data2/dataset/chenwy/21164-data/dpo_dataset/u2net_next_inpainting/HPDv3/pickscore/pickscore_score.csv"

# all_information_csv_file_path = "/data_center/data2/dataset/chenwy/21164-data/dpo_dataset/u2net_next_inpainting/pick-a-pic-v2-top_4096/top_4096_images.csv"
# pickscore_csv_file_path = "/data_center/data2/dataset/chenwy/21164-data/dpo_dataset/u2net_next_inpainting/pick-a-pic-v2-top_4096/pickscore/pickscore_score.csv"
# anime_csv_file_path = "/data_center/data2/dataset/chenwy/21164-data/dpo_dataset/u2net_next_inpainting/HPDv3/qwen3_anime/qwen3_anime.csv"
# colorfulness_csv_file_path = "/data_center/data2/dataset/chenwy/21164-data/dpo_dataset/u2net_next_inpainting/HPDv3/colorfulness/colorfulness_score.csv"

pickscore_df = pd.read_csv(pickscore_csv_file_path, dtype={"uid": str})
all_information_df = pd.read_csv(all_information_csv_file_path, dtype={"uid": str})
anime_df = pd.read_csv(anime_csv_file_path, dtype={"uid": str})
colorfulness_df = pd.read_csv(colorfulness_csv_file_path, dtype={"uid": str})

# valid_pickscore_uids = pickscore_df[pickscore_df['real_image_score'] - pickscore_df['fake_image_score'] > 0.02]['uid'].tolist()
valid_real_image_uids = anime_df[anime_df['anime'] == 'no']['uid'].tolist()
colorfulness_df = colorfulness_df[colorfulness_df['uid'].isin(valid_real_image_uids)]
avg_colorfulness = colorfulness_df['real_image_score'].mean()
valid_colorfulness_uids = colorfulness_df[colorfulness_df['real_image_score'] > avg_colorfulness]['uid'].tolist()

valid_uids = list(set(valid_real_image_uids) & set(valid_colorfulness_uids))
print(f"valid_uids: {len(valid_uids)}")
print(f"valid_real_image_uids: {len(valid_real_image_uids)} / {len(anime_df)}")
print(f"valid_colorfulness_uids: {len(valid_colorfulness_uids)} / {len(colorfulness_df)}; avg_colorfulness: {avg_colorfulness:.4f}")
# print(f"valid_pickscore_uids: {len(valid_pickscore_uids)} / {len(pickscore_df)}")

# 从 valid_uids 中筛选出 pickscore real_image_score 值最高的前 N 张图片
# 可配置参数：选择前 N 个样本
TOP_N = 512

# Step 1: 从 pickscore_df 中筛选出 uid 在 valid_uids 中的行
pickscore_valid_df = pickscore_df[pickscore_df['uid'].isin(valid_uids)].copy()
print(f"\nStep 1: Pickscore samples in valid_uids: {len(pickscore_valid_df)}")

# Step 2: 按照 real_image_score 降序排序
pickscore_valid_df = pickscore_valid_df.sort_values('real_image_score', ascending=False)

# Step 3: 取前 N 个
top_N_df = pickscore_valid_df.head(TOP_N)
print(f"Step 2: Top {TOP_N} samples selected")
print(f"  Real image score range: {top_N_df['real_image_score'].min():.4f} - {top_N_df['real_image_score'].max():.4f}")
print(f"  Score gap range: {(top_N_df['real_image_score'] - top_N_df['fake_image_score']).min():.4f} - {(top_N_df['real_image_score'] - top_N_df['fake_image_score']).max():.4f}")

# 获取 top N 的 uid 列表
top_N_uids = top_N_df['uid'].tolist()
print(f"\nTop {TOP_N} UIDs selected: {len(top_N_uids)}")

# Step 4: Generate CSV file with image paths
ext_list = [".png", ".PNG", ".jpg", ".jpeg", ".JPG", ".JPEG"]
real_image_dir = "/data_center/data2/dataset/chenwy/21164-data/dpo_dataset/u2net_next_inpainting/HPDv3/real"
fake_image_dir = "/data_center/data2/dataset/chenwy/21164-data/dpo_dataset/u2net_next_inpainting/HPDv3/fake"
# mask_image_dir = "/data_center/data2/dataset/chenwy/21164-data/dpo_dataset/u2net_next_inpainting/HPDv3/mask"
final_information_list = []
for index, row in top_N_df.iterrows():
    uid = row["uid"]
    win_image_path, lose_image_path, mask_image_path = None, None, None
    
    for ext in ext_list:
        win_image_path = os.path.join(real_image_dir, f"{uid}{ext}")
        if os.path.exists(win_image_path):
            break
        win_image_path = None
        
    if win_image_path is None:
        print(f"Warning: No valid image found for uid {uid}")
        continue
    
    for ext in ext_list:
        lose_image_path = os.path.join(fake_image_dir, f"{uid}{ext}")
        if os.path.exists(lose_image_path):
            break
        lose_image_path = None
        
    if lose_image_path is None:
        print(f"Warning: No valid image found for uid {uid}")
        continue
    
    # for ext in ext_list:
    #     mask_image_path = os.path.join(mask_image_dir, f"{uid}{ext}")
    #     if os.path.exists(mask_image_path):
    #         break
    # if mask_image_path is None:
    #     print(f"Warning: No valid mask image found for uid {uid}")
    #     continue
    
    prompt = all_information_df.loc[all_information_df['uid'] == uid, 'prompt'].values[0]
    final_information_list.append({
        'uid': uid,
        'prompt': prompt,
        'win_image_path': win_image_path,
        'lose_image_path': lose_image_path,
        # 'mask_image_path': mask_image_path
    })

print(f"\nFinal valid samples: {len(final_information_list)}")
final_output_path = f"/data_center/data2/dataset/chenwy/21164-data/dpo_dataset/u2net_next_inpainting/HPDv3/top_{TOP_N}_images_no_anime_colorfulness_pickscore-hpdv3_all_uids.csv"
pd.DataFrame(final_information_list).to_csv(final_output_path, index=False)
print(f"Saved to: {final_output_path}")
```

```python
#### Link win image and lose image ####
import os
import pandas as pd

csv_file_name = "top_512_images_no_anime_colorfulness_pickscore_0.02-hpdv3_all-uids"
base_dir = "/data_center/data2/dataset/chenwy/21164-data/dpo_dataset/u2net_next_inpainting/civitai-top-sfw-images-with-metadata"
base_target_image_dir = os.path.join(base_dir, "visualization")
os.makedirs(base_target_image_dir, exist_ok=True)

ext_list = [".png", ".PNG", ".jpg", ".jpeg", ".JPG", ".JPEG"]
df = pd.read_csv(os.path.join(base_dir, f"{csv_file_name}.csv"))
target_image_dir = os.path.join(base_target_image_dir, csv_file_name)
os.makedirs(os.path.join(target_image_dir, "real"), exist_ok=True)
os.makedirs(os.path.join(target_image_dir, "fake"), exist_ok=True)

for index, row in df.iterrows():
    uid = row['uid']
    source_win_image_path = row['win_image_path']
    source_lose_image_path = row['lose_image_path']

    win_image_name = os.path.basename(source_win_image_path)
    lose_image_name = os.path.basename(source_lose_image_path)

    target_win_image_path = os.path.join(target_image_dir, "real", win_image_name)
    target_lose_image_path = os.path.join(target_image_dir, "fake", lose_image_name)

    os.symlink(source_win_image_path, target_win_image_path)
    os.symlink(source_lose_image_path, target_lose_image_path)
    
    mask_path = os.path.join(base_dir, "mask", f"{uid}.png")
    if os.path.exists(mask_path):
        target_mask_path = os.path.join(target_image_dir, "mask", os.path.basename(mask_path))
        os.makedirs(os.path.dirname(target_mask_path), exist_ok=True)
        os.symlink(mask_path, target_mask_path)
```
