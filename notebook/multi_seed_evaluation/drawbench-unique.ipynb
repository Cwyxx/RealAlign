```python

base_txt_path = "/data3/chenweiyan/notebook/fine-tune-diffusion/spo_gitee/DiffusionNFT/dataset/drawbench/test.txt"
target_txt_path = "/data3/chenweiyan/notebook/fine-tune-diffusion/spo_gitee/DiffusionNFT/dataset/drawbench-unique/test.txt"

import os
os.makedirs(os.path.dirname(target_txt_path), exist_ok=True)

if os.path.exists(base_txt_path):
    file_to_read = base_txt_path
    print(f"读取目标文件: {base_txt_path}")
else:
    file_to_read = base_txt_path
    print(f"目标文件不存在，从源文件读取: {base_txt_path}")

with open(file_to_read, 'r', encoding='utf-8') as f:
    lines = f.readlines()

seen = set()
unique_lines = []
for line in lines:
    line_stripped = line.strip()
    if line_stripped and line_stripped not in seen:
        seen.add(line_stripped)
        unique_lines.append(line.rstrip('\n\r') + '\n')  # 保留原始换行符格式

with open(target_txt_path, 'w', encoding='utf-8') as f:
    f.writelines(unique_lines)

print(f"去重完成！原始行数: {len(lines)}, 去重后行数: {len(unique_lines)}, 删除重复: {len(lines) - len(unique_lines)} 行")
```
