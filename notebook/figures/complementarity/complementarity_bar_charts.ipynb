```python
#### SD-1.5 ####

import matplotlib as mpl
mpl.rcParams['hatch.linewidth'] = 2.0
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick
import numpy as np

#### Pick-a-Pic-v2 ####
# # 数据准备
# data = {
#     "SD-1.5": { "PickScore": 20.65, "ImageReward": 0.16, "UnifiedReward": 2.92, "HPSv3": 5.98, "DeQA": 3.70, "Aesthetic": 5.48},
#     "SPO": { "PickScore": 21.19, "ImageReward": 0.33, "UnifiedReward": 2.98, "HPSv3": 7.31, "DeQA": 3.40, "Aesthetic": 5.76},
#     "Diffusion-DPO": { "PickScore": 21.03, "ImageReward": 0.33, "UnifiedReward": 3.03, "HPSv3": 6.80, "DeQA": 3.78, "Aesthetic": 5.59},
#     "Ours": { "PickScore": 21.04, "ImageReward": 0.38, "UnifiedReward": 3.06, "HPSv3": 7.33, "DeQA": 3.96, "Aesthetic": 5.64},
#     "Diffusion-DPO+Ours": { "PickScore": 21.19, "ImageReward": 0.42, "UnifiedReward": 3.08, "HPSv3": 7.50, "DeQA": 3.95, "Aesthetic": 5.65},
# }
#### Pick-a-Pic-v2 ####

### Parti-Prompts ####
# 数据准备
data = {
    "SD-1.5": { "PickScore": 21.34, "ImageReward": 0.25, "UnifiedReward": 3.15, "HPSv3": 5.69, "DeQA": 3.70, "Aesthetic": 5.39},
    "SPO": { "PickScore": 21.70, "ImageReward": 0.46, "UnifiedReward": 3.21, "HPSv3": 7.02, "DeQA": 3.39, "Aesthetic": 5.67},
    "Diffusion-DPO": { "PickScore": 21.58, "ImageReward": 0.40, "UnifiedReward": 3.25, "HPSv3": 6.48, "DeQA": 3.78, "Aesthetic": 5.47},
    "Ours": { "PickScore": 21.64, "ImageReward": 0.41, "UnifiedReward": 3.28, "HPSv3": 7.06, "DeQA": 3.96, "Aesthetic": 5.49},
    "Diffusion-DPO+Ours": { "PickScore": 21.73, "ImageReward": 0.47, "UnifiedReward": 3.31, "HPSv3": 7.29, "DeQA": 3.94, "Aesthetic": 5.51},
}
### Parti-Prompts ####


methods = list(data.keys())
metrics = list(data[methods[0]].keys())
formatted_methods = []
for m in methods:
    if m == "Diffusion-DPO+Ours":
        formatted_methods.append(r"Diffusion-DPO$\mathbf{+Ours}$")
    elif m == "Ours":
        formatted_methods.append(r"$\mathbf{Ours}$")
    else:
        formatted_methods.append(m)

# 宫崎骏 RGB 配色
rgb_colors = [
    (70/255, 120/255, 142/255), (120/255, 183/255, 201/255), 
    (246/255, 224/255, 147/255), (229/255, 139/255, 123/255), 
    (229/255, 139/255, 123/255) # (151/255, 179/255, 25/255)
]

# 字体配置
title_font_size = 24
x_tick_font_size = 22
y_tick_font_size = 16
value_font_size = 16

fig, axes = plt.subplots(1, 6, figsize=(30, 7), dpi=300)

for i, metric in enumerate[str](metrics):
    ax = axes[i]
    values = [data[method][metric] for method in methods]
    
    # 绘制柱状图
    bars = ax.bar(methods, values, color=rgb_colors, edgecolor='white', linewidth=1.2)
    
    # --- 统一保留两位小数 ---
    # --- 遍历柱子应用特殊标记 ---
    for j, bar in enumerate(bars):
        method_name = methods[j]
        
        if method_name == "Diffusion-DPO+Ours":
            bar.set_hatch('/')
            bar.set_edgecolor('white')
        
        height = bar.get_height()
        ax.annotate(f'{height:.2f}',
                    xy=(bar.get_x() + bar.get_width() / 2, height),
                    xytext=(0, 5), 
                    textcoords="offset points",
                    ha='center', va='bottom', 
                    fontsize=value_font_size, fontweight='bold', color='#333333')

    # 设置 y 轴保留一位小数
    ax.yaxis.set_major_formatter(mtick.FormatStrFormatter('%.1f'))
    ax.yaxis.set_major_locator(plt.MaxNLocator(nbins=4))
    
    # 标题加大并带箭头
    # ax.set_title(f"{metric}$\\uparrow$", fontsize=title_font_size, pad=10)
    ax.set_ylabel(f"{metric} $\\uparrow$", fontsize=title_font_size, labelpad=5)
    
    # 轴标签调整
    ax.tick_params(axis='x', labelsize=x_tick_font_size)
    ax.tick_params(axis='y', labelsize=y_tick_font_size)
    ax.set_xticklabels(formatted_methods, rotation=40, ha='right')
    
    # 优化显示范围：为顶部数字留出 30% 的额外空间
    min_v, max_v = min(values), max(values)
    y_range = max_v - min_v
    ax.set_ylim(max(0, min_v - y_range * 0.2), max_v + y_range * 0.3)
    
    # 细节美化
    # ax.yaxis.grid(True, linestyle='--', alpha=0.3)
    # ax.spines['top'].set_visible(False)
    # ax.spines['right'].set_visible(False)
    ax.set_axisbelow(True)

plt.tight_layout()
plt.savefig('sd15_partiprompts_bar_chart.svg', dpi=300, bbox_inches='tight')
plt.show()
```
