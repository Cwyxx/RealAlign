```python
#### User Study ####

import matplotlib as mpl
mpl.rcParams['hatch.linewidth'] = 2.0
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import numpy as np

# 数据准备：Win / Tie / Lose (%)
results = [
    {"title": "vs. SD-3.5-M",       "win": 63.33, "tie": 16.67, "lose": 20.00},
    {"title": "vs. Diffusion-DPO",  "win": 56.67, "tie": 10.00, "lose": 33.33},
]

# # 配色 - v1
# color_lose = '#316394'   # Lose：深蓝
# color_tie  = '#73bdd6'   # Tie：浅蓝
# color_win  = '#ffe7b5'   # Win：黄

# # 配色 - v2
# color_lose = '#2c7fb9'   # Lose：深蓝
# color_tie  = '#41b6c5'   # Tie：浅蓝
# color_win  = '#a2dab5'   # Win：绿色

# # 配色 - v3
# color_lose = '#b8dfee'   # Lose：浅蓝
# color_tie  = '#ffffc1'   # Tie：浅黄
# color_win  = '#ffc287' # '#feba79'   # Win：绿色

# # 配色 - v4
color_lose = '#2c7fb9'   # Lose：浅蓝
color_tie  = '#60c0cd'   # Tie：
color_win  = '#b1e0c0'   # Win：浅绿


# 字体配置
title_font_size  = 20
tick_font_size   = 18
value_font_size  = 20
legend_font_size = 20

fig, axes = plt.subplots(1, 2, figsize=(16, 1.6), dpi=300)

bar_h = 0.28

for ax, item in zip(axes, results):
    lose, tie, win = item["lose"], item["tie"], item["win"]

    # # 顺序：Lose → Tie → Win（从左到右堆叠在 y=0 这一根条上）
    # ax.barh(0, lose,                        height=bar_h, color=color_lose, edgecolor='black', linewidth=0.8)
    # ax.barh(0, tie,  left=lose,             height=bar_h, color=color_tie,  edgecolor='black', linewidth=0.8)
    # ax.barh(0, win,  left=lose + tie,       height=bar_h, color=color_win,  edgecolor='black', linewidth=0.8)

    # # 段中文字
    # ax.text(lose / 2,                 0, f'{lose:.0f}%',
    #         ha='center', va='center', fontsize=value_font_size, color='white')
    # ax.text(lose + tie / 2,           0, f'{tie:.0f}%',
    #         ha='center', va='center', fontsize=value_font_size, color='#222222')
    # ax.text(lose + tie + win / 2,     0, f'{win:.0f}%',
    #         ha='center', va='center', fontsize=value_font_size, color='#222222')
    
    # 顺序：Win → Tie → Lose
    ax.barh(0, win,                         height=bar_h, color=color_win,  alpha=0.8, edgecolor='black', linewidth=0.8)
    ax.barh(0, tie,  left=win,              height=bar_h, color=color_tie,  edgecolor='black', linewidth=0.8)
    ax.barh(0, lose, left=win + tie,        height=bar_h, color=color_lose, edgecolor='black', linewidth=0.8)

    # 段中文字
    ax.text(win / 2,                 0, f'{win:.0f}%',
            ha='center', va='center', fontsize=value_font_size, color='#222222')

    ax.text(win + tie / 2,           0, f'{tie:.0f}%',
            ha='center', va='center', fontsize=value_font_size, color='#222222')

    ax.text(win + tie + lose / 2,    0, f'{lose:.0f}%',
            ha='center', va='center', fontsize=value_font_size, color='#F2F2F2')

    # x 轴
    ax.set_xlim(0, 100)
    ax.set_xticks([0, 20, 40, 60, 80, 100])
    ax.tick_params(axis='x', labelsize=tick_font_size)
    ax.xaxis.set_minor_locator(mpl.ticker.MultipleLocator(5))
    ax.tick_params(axis='x', which='minor',
               direction='out', length=3, width=0.8)

    # y 轴隐藏
    ax.set_yticks([])
    ax.set_ylim(-0.18, 0.18)

    # # 仅保留底部边框
    # for side in ['top', 'right', 'left']:
    #     ax.spines[side].set_visible(False)
    for side in ['top', 'right', 'left', 'bottom']:
        ax.spines[side].set_visible(True)
        # ax.spines[side].set_linewidth(1.0)
        ax.spines[side].set_color('black')

    # 标题置于下方
    ax.set_xlabel(item["title"], fontsize=title_font_size, labelpad=8)

# 共享图例（顺序：Win-Tie-Lose，更符合阅读习惯）
legend_handles = [
    mpatches.Patch(facecolor=color_win,  edgecolor='black', label='Win'),
    mpatches.Patch(facecolor=color_tie,  edgecolor='black', label='Tie'),
    mpatches.Patch(facecolor=color_lose, edgecolor='black', label='Lose'),
]
fig.legend(handles=legend_handles,
           loc='upper center', bbox_to_anchor=(0.5, 1.33),
           ncol=3, frameon=False, fontsize=legend_font_size,
           handlelength=2.6, handleheight=1.3, handletextpad=0.6, columnspacing=2.5)

plt.tight_layout()
plt.savefig('../user_study/user_study.svg', dpi=300, bbox_inches='tight')
plt.savefig('../user_study/user_study.pdf', dpi=300, bbox_inches='tight')
plt.show()
```
