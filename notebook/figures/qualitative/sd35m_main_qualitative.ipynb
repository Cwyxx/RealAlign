```python
### SD-3.5-Medium-visualization-v3 ###
import os
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from matplotlib.font_manager import FontProperties

# 字体配置
comic_font = FontProperties(
    fname="/data3/chenweiyan/notebook/fine-tune-diffusion/spo_gitee/notebook/visualization/Comic Sans MS.ttf",
    size=14
)

base_image_dir = f"/data_center/data2/dataset/chenwy/21164-data/diffusion-dpo/sd-3-5-medium"

# 配置列表
method_list = [
    "sd-3-5-medium",
    "FlowGRPO-PickScore",
    "irl_top_512_images_no_anime_colorfulness_pickscore_0.02-hpdv3_all_lr_0.0002_ckpt_3200-dpo_top_512_images_no_anime_colorfulness_pickscore_0.02-hpdv3_all",
    "FlowGRPO-PickScore-next-top_512_images_no_anime_colorfulness_pickscore_002-hpdv3_all-inpainting"
]
method_name_list = ["SD-3.5-M", "FlowGRPO", "Ours", "FlowGRPO+Ours"] # 
ckpt_list = [0, 0, 450, 300]

dataset_name_list = [ "partiprompts", "partiprompts", "HPDv2-anime"] # "partiprompts",
seed_list = [42, 42, 42] # 42
image_id_list = [ 243, 264, 9] # 1175, 

# --- 逻辑交换：现在行是样本，列是方法 ---
num_rows = len(image_id_list)    # 样本数 -> 行
num_columns = len(method_list)   # 方法数 -> 列

fig, axes = plt.subplots(num_rows, num_columns, figsize=(2 * num_columns, 2 * num_rows), dpi=300)

# 确保 axes 始终是二维数组
if num_rows == 1:
    axes = [axes]
if num_columns == 1:
    axes = [[ax] for ax in axes]

# --- 核心绘制逻辑 ---
# 外层循环遍历行 (样本)
for row_idx, (seed, image_id, dataset_name) in enumerate(zip(seed_list, image_id_list, dataset_name_list)):
    # 内层循环遍历列 (方法)
    for col_idx, (method, ckpt, method_name) in enumerate(zip(method_list, ckpt_list, method_name_list)):
        
        image_path = os.path.join(base_image_dir, f"generate_images_seed_{seed}/{dataset_name}", method, f"ckpt-{ckpt}", "images", f"{image_id:05d}.png")
        ax = axes[row_idx][col_idx]
        
        try:
            img = mpimg.imread(image_path)
            ax.imshow(img, aspect="auto")
        except FileNotFoundError:
            ax.text(0.5, 0.5, "Image Not Found", ha='center', va='center', fontsize=8)
        
        ax.set_xticks([])
        ax.set_yticks([])
        # 隐藏边框
        for spine in ax.spines.values():
            spine.set_visible(False)
            
        # 在第一行设置顶部标题 (方法名)
        if row_idx == 0:
            ax.set_title(method_name, fontproperties=comic_font, pad=10)
            
        # # 在第一列设置左侧标签 (样本信息)
        # if col_idx == 0:
        #     ax.set_ylabel(
        #         f"{dataset_name}\nID: {image_id}", 
        #         fontproperties=comic_font, 
        #         labelpad=10,
        #         rotation=90,
        #         va='center'
        #     )

# 调整间距，wspace和hspace设为较小值可以紧凑排布
plt.subplots_adjust(wspace=0, hspace=0)
plt.savefig("sd35m_main_qualitative_alt.pdf", format="pdf", bbox_inches="tight", dpi=300)
plt.show()
```
