```python
### SD-1.5 main qualitative visualization ###
import os
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from matplotlib.font_manager import FontProperties

# 字体配置
comic_font = FontProperties(
    fname="/data3/chenweiyan/notebook/fine-tune-diffusion/spo_gitee/notebook/visualization/Comic Sans MS.ttf",
    size=14
)

base_image_dir = f"/data_center/data2/dataset/chenwy/21164-data/diffusion-dpo/sd-v1-5"

# 配置列表
method_list = [
    "sd-v1-5",
    "dpo-official",
    "irl_top_512_images_no_anime_colorfulness_pickscore_0.02-hpdv3_all_ckpt_1600-dpo_2000_top_512_images_no_anime_colorfulness_pickscore_0.02-hpdv3_all_inpainting",
    "dpo_official-irl_top_512_images_no_anime_colorfulness_pickscore_0.02-hpdv3_all_ckpt_1600-dpo_2000_top_512_images_no_anime_colorfulness_pickscore_0.02-hpdv3_all-inpainting"
]
method_name_list = ["SD-1.5", "Diffusion-DPO", "Ours", "Diffusion-DPO+Ours"]
ckpt_list = [0, 0, 800, 800]

dataset_name_list = [ "partiprompts", "pick_a_pic_v2", "partiprompts" ] # "pick_a_pic_v2",
seed_list = [ 42, 123, 42 ] # 42,
image_id_list = [ 238, 279, 940] # 357, 

# --- 逻辑交换：现在行是样本，列是方法 ---
num_rows = len(image_id_list)   # 样本数 -> 行
num_columns = len(method_list)  # 方法数 -> 列

fig, axes = plt.subplots(num_rows, num_columns, figsize=(2 * num_columns, 2 * num_rows), dpi=300)

# 确保 axes 始终是二维数组
if num_rows == 1:
    axes = [axes]
if num_columns == 1:
    axes = [[ax] for ax in axes]

# --- 核心绘制逻辑 ---
# 外层循环遍历行 (样本)
for row_idx, (seed, image_id, dataset_name) in enumerate(zip(seed_list, image_id_list, dataset_name_list)):
    # 内层循环遍历列 (方法)
    for col_idx, (method, ckpt, method_name) in enumerate(zip(method_list, ckpt_list, method_name_list)):
        
        image_path = os.path.join(base_image_dir, f"generate_images_seed_{seed}/{dataset_name}", method, f"ckpt-{ckpt}", "images", f"{image_id:05d}.png")
        
        ax = axes[row_idx][col_idx]
        
        try:
            img = mpimg.imread(image_path)
            ax.imshow(img, aspect="auto")
        except FileNotFoundError:
            ax.text(0.5, 0.5, "NF", ha='center', va='center') # 如果路径不对，显示 Not Found
        
        ax.set_xticks([])
        ax.set_yticks([])
        for spine in ax.spines.values():
            spine.set_visible(False)
            
        # 在第一行设置 Title，标注方法名
        if row_idx == 0:
            ax.set_title(method_name, fontproperties=comic_font, pad=10)
            
        # # 在第一列设置 ylabel，标注样本 ID (可选)
        # if col_idx == 0:
        #     ax.set_ylabel(f"ID: {image_id}", fontproperties=comic_font, labelpad=10)

plt.subplots_adjust(wspace=0, hspace=0) # 留一点极小的缝隙通常看起来更整洁
plt.savefig("sd15_main_qualitative.pdf", format="pdf", bbox_inches="tight", dpi=300)
plt.show()
```
