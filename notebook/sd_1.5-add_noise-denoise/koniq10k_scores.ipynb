```python
#### Static Information ####
import os
import pandas as pd
import matplotlib.pyplot as plt

csv_file_path = "/data_center/data2/dataset/chenwy/21164-data/dpo_dataset/konIQ-10k/koniq10k_scores_and_distributions.csv"
df = pd.read_csv(csv_file_path)

plt.figure(figsize=(10, 6), dpi=300)
plt.hist(df['MOS'], bins=50, edgecolor='black', alpha=0.7)
plt.xlabel('MOS value', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.grid(True, alpha=0.3)
plt.axvline(df['MOS'].mean(), color='red', linestyle='--', linewidth=2, label=f'mean: {df["MOS"].mean():.2f}')
plt.axvline(df['MOS'].median(), color='green', linestyle='--', linewidth=2, label=f'median: {df["MOS"].median():.2f}')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
#### remove low quality images ####
from tqdm import tqdm
image_dir = "/data_center/data2/dataset/chenwy/21164-data/dpo_dataset/konIQ-10k/1024x768"
df = df[df['MOS'] >= 3.5]

for image_name in tqdm(os.listdir(image_dir)):
    row = df[df['image_name'] == image_name]
    if len(row) > 0:
        continue
    
    image_path = os.path.join(image_dir, image_name)
    if os.path.exists(image_path):
        os.remove(image_path)
```
