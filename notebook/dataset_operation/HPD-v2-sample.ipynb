```python
import os
import random
import json

seed = 42
hpdv2_type_list=["anime", "concept-art", "paintings", "photo"]
base_dir="/data3/chenweiyan/notebook/fine-tune-diffusion/spo_gitee/DiffusionNFT/dataset"
base_hpdv2_dir=os.path.join(base_dir, "HPDv2")

for hpdv2_type in hpdv2_type_list:
    hpdv2_type_json = os.path.join(base_hpdv2_dir, f"{hpdv2_type}.json")
    target_txt_path = os.path.join(base_dir, f"HPDv2-{hpdv2_type}", "test.txt")
    with open(hpdv2_type_json, "r") as f:
        hpdv2_data = json.load(f)
    
    random.seed(seed)
    
    # 提取所有 prompt
    # 如果 hpdv2_data 是字符串列表，直接使用
    # 如果是对象列表，提取 prompt 字段
    if isinstance(hpdv2_data, list):
        if len(hpdv2_data) > 0 and isinstance(hpdv2_data[0], str):
            prompts = [p.strip() for p in hpdv2_data if p.strip()]
        elif len(hpdv2_data) > 0 and isinstance(hpdv2_data[0], dict):
            prompts = [entry.get('prompt', '').strip() for entry in hpdv2_data if entry.get('prompt', '').strip()]
        else:
            prompts = []
    else:
        prompts = []
    
    # 随机选择 200 条（如果数据少于 200 条，则选择全部）
    sample_size = min(200, len(prompts))
    selected_prompts = random.sample(prompts, sample_size)
    
    # 确保目标目录存在
    target_dir = os.path.dirname(target_txt_path)
    os.makedirs(target_dir, exist_ok=True)
    
    # 保存到文件，每行一个 prompt
    with open(target_txt_path, "w", encoding="utf-8") as f:
        for prompt in selected_prompts:
            f.write(prompt + "\n")
    
    print(f"{hpdv2_type}: 从 {len(prompts)} 条中随机选择了 {sample_size} 条，已保存到 {target_txt_path}")
```
