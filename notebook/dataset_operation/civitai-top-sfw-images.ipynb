```python
### transform civitai-top-sfw-images.jsonl to csv ###
import json
import pandas as pd
import os
import csv

prompt_json_path = "/data_center/data2/dataset/chenwy/21164-data/dpo_dataset/civitai-top-sfw-images-with-metadata/prompts.json"

with open(prompt_json_path, 'r', encoding='utf-8') as f:
    data = json.load(f)

rows = []
for filename, content in data.items():
    uid = os.path.splitext(filename)[0]
    prompt = content.get('prompt', '')
    rows.append({'uid': uid, 'prompt': prompt})

df = pd.DataFrame(rows)
output_csv_path = "/data_center/data2/dataset/chenwy/21164-data/dpo_dataset/civitai-top-sfw-images-with-metadata/uid_prompt.csv"

# 使用 quoting=csv.QUOTE_ALL 确保所有字段都被引号包裹，避免特殊字符导致解析错误
df.to_csv(
    output_csv_path, 
    index=False, 
    encoding='utf-8',
    quoting=csv.QUOTE_ALL,  # 所有字段都用引号包裹
    escapechar='\\',  # 转义字符
    doublequote=True  # 引号内的引号用双引号表示
)

print(f"转换完成！共 {len(df)} 条记录")
print(f"CSV 文件已保存到: {output_csv_path}")
print(f"\n前5条数据预览:")
print(df.head())
```
