```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter

method_list = [
    "SD-3.5-M",
    "AFPO (HPDv3)",
    "AFPO (Civitai-top)",
    "FlowGRPO",
    "GRPO-Guard",
    "DiffusionNFT",
]

human_preference_dict = {
    "SD-3.5-M": 0,
    "FlowGRPO": 0.837719295,
    "GRPO-Guard": 0.8824561,
    "DiffusionNFT": 0.92807,
    "AFPO (HPDv3)": 0.69122805,
    "AFPO (Civitai-top)": 0.85526,
}
oneig_benchmark_dict = {
    "SD-3.5-M": 0.357,
    "FlowGRPO": 0.260,
    "GRPO-Guard": 0.241,
    "DiffusionNFT": 0.175,
    "AFPO (HPDv3)": 0.321,
    "AFPO (Civitai-top)": 0.334,
}

# ---------- figure style ----------
plt.rcParams.update({
    "axes.labelsize": 14,
    "xtick.labelsize": 8,
    "ytick.labelsize": 8,
    "legend.fontsize": 8,
    # "font.family": "DejaVu Sans",
})

fig, ax = plt.subplots(figsize=(5, 4), dpi=300)

markers_default = ["o", "s", "^", "D", "P", "X"]
colors_default = [ "#A4C6E2", "#E4A39D", "skip",  "#AFD498", "#AE9DC7", "#e7dbd3"]

# Plot points
for i, m in enumerate(method_list):
    x = human_preference_dict[m]
    y = oneig_benchmark_dict[m]

    # --- 样式逻辑 ---
    if m == "AFPO (Civitai-top)":
        mk = "*"
        c = "#EF9963"
        s = 320
        zorder = 10
        lw = 1.0
        alpha = 1.0
    else:
        mk = markers_default[i]
        c = colors_default[i]
        zorder = 3
        lw = 0.8
        alpha = 1.0
        s = 160
        
        if m == "FlowGRPO":
            s = 96
            # alpha = 0.9
        elif m == "AFPO (HPDv3)":
            s = 120

    ax.scatter(
        x, y,
        s=s,
        marker=mk,
        facecolors=c,
        edgecolors="black",
        linewidths=lw,
        zorder=zorder,
        label=m,
        alpha=alpha
    )

# ---------- 智能标签布局 ----------
label_offsets = {
    "SD-3.5-M": (10, 0),
    "FlowGRPO": (-8, -5),            
    "GRPO-Guard": (10, -2),
    "DiffusionNFT": (-8, 5),
    "AFPO (HPDv3)": (-10, 0),
    "AFPO (Civitai-top)": (0, 18),
}

for m in method_list:
    x = human_preference_dict[m]
    y = oneig_benchmark_dict[m]
    dx, dy = label_offsets.get(m, (5, 5))
    
    if dx < 0: ha = 'right'
    elif dx > 0: ha = 'left'
    else: ha = 'center'
    
    # 字体微调：主角稍微大一点点
    font_size = 10
    font_weight = "normal"
    text_color = "#333333"

    ax.annotate(
        m, (x, y),
        xytext=(dx, dy), textcoords="offset points",
        fontsize=font_size, 
        color=text_color,
        fontweight=font_weight,
        ha=ha, va='center', zorder=20
    )

# Axes labels
ax.set_xlabel(r"Human Preference$\uparrow$")
ax.set_ylabel(r"Stylization$\uparrow$")

# Grid + spines
ax.grid(True, linestyle="--", linewidth=0.8, alpha=0.3, zorder=0)
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)

# Ticks
ax.set_xlim(-0.1, 1.1)
ax.set_xticks(np.arange(0.0, 1.1, 0.2))
ax.set_ylim(0.14, 0.41)
ax.set_yticks(np.arange(0.15, 0.41, 0.05))

ax.xaxis.set_major_formatter(FuncFormatter(lambda v, pos: f"{v:.1f}"))
ax.yaxis.set_major_formatter(FuncFormatter(lambda v, pos: f"{v:.2f}"))

# ---------- 优化后的图例 (Legend) ----------
legend = ax.legend(
    loc="lower left",
    frameon=True,         # 显示边框
    fancybox=True,        # 圆角边框（看起来柔和一点）
    edgecolor="gray",     # 边框颜色（灰色比黑色更精致）
    framealpha=0.3,       # 背景稍微一点点透，融合感更好
    borderpad=0.5,        # 【关键】内边距：增加边框和文字之间的距离
    labelspacing=0.8,     # 行间距：增加每一行之间的距离
    handletextpad=0.5,    # 图标和文字之间的距离
    borderaxespad=0.8     # 图例和坐标轴之间的距离
)

# 统一调整图例中图标的大小（可选，让图例看起来更整齐）
for handle in legend.legend_handles:
    if handle.get_label() == "AFPO (Civitai-top)":
        handle.set_sizes([180.0]) 
    elif handle.get_label() == "AFPO (HPDv3)":
        handle.set_sizes([80.0]) 
    elif handle.get_label() == "FlowGRPO":
        handle.set_sizes([75.0]) 
    else:
        handle.set_sizes([100.0]) 

fig.tight_layout()
fig.savefig("2026-ICML-human_preference-vs-oneig_style.pdf", dpi=300,bbox_inches="tight")
plt.show()
```
