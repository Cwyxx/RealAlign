```python
import random
random.seed(456)
dataset_name = "HPDv2-photo"
txt_path = f"/data3/chenweiyan/notebook/fine-tune-diffusion/spo_gitee/DiffusionNFT/dataset/{dataset_name}/test.txt"

# 读取 test.txt 文件，保留行号信息
with open(txt_path, 'r', encoding='utf-8') as f:
    # 保存 (行号, prompt) 的元组列表
    all_prompts_with_line = [(i, line.strip()) for i, line in enumerate(f.readlines())]

# 随机挑选 15 条 prompt（包含行号）
selected_prompts = random.sample(all_prompts_with_line, 100)

# 按行号排序
selected_prompts_sorted = sorted(selected_prompts, key=lambda x: x[0])

# 打印选中的 prompts
print(f"总共有 {len(all_prompts_with_line)} 条 prompts")
print(f"\n随机挑选的 15 条 prompts（按行号排序）:\n")
for i, (line_num, prompt) in enumerate(selected_prompts_sorted, 1):
    print(f"{i}. [行号 {line_num}] {prompt}")
```
