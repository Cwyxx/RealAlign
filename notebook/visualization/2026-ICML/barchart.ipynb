```python
#### Inpainting mask strategy SD-1.5 Improvement ####

import matplotlib as mpl
mpl.rcParams['hatch.linewidth'] = 2.0
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick
import numpy as np

#### Pick-a-Pic-v2 ####
# 数据准备
data = {
    "SAM-random_top_5" : {
        "PickScore": 0.06, 
        "ImageReward": 0.01,
        "UnifiedReward": 0.02,
        "HPSv3": 0.27,
        "DeQA": 0.11,
        "Aesthetic": 0.03
    },
    "Full-mask" : {
        "PickScore": 0.26,
        "ImageReward": 0.05,
        "UnifiedReward": 0.05,
        "HPSv3": 0.20,
        "DeQA": 0.21,
        "Aesthetic": 0.10
    },
    "Text-to-Image" : {
        "PickScore": 0.16,
        "ImageReward": 0.01,
        "UnifiedReward": 0.02,
        "HPSv3": 0.28,
        "DeQA": 0.21,
        "Aesthetic": 0.07
    },
    "Saliency-mask" : {
        "PickScore": 0.18,
        "ImageReward": 0.04,
        "UnifiedReward": 0.06,
        "HPSv3": 0.49,
        "DeQA": 0.21,
        "Aesthetic": 0.07
    }
}
#### Pick-a-Pic-v2 ####



methods = list(data.keys())
metrics = list(data[methods[0]].keys())
formatted_methods = []
for m in methods:
    formatted_methods.append(m)

# 宫崎骏 RGB 配色
rgb_colors = [
    (70/255, 120/255, 142/255), 
    (120/255, 183/255, 201/255), 
    (246/255, 224/255, 147/255), 
    (229/255, 139/255, 123/255), 
    # (229/255, 139/255, 123/255) # (151/255, 179/255, 25/255)
]

# 字体配置
title_font_size = 24
x_tick_font_size = 22
y_tick_font_size = 16
value_font_size = 16

fig, axes = plt.subplots(1, 6, figsize=(30, 7), dpi=300)

for i, metric in enumerate[str](metrics):
    ax = axes[i]
    values = [data[method][metric] for method in methods]
    
    # 绘制柱状图
    bars = ax.bar(methods, values, color=rgb_colors, edgecolor='white', linewidth=1.2)
    
    # --- 统一保留两位小数 ---
    # --- 遍历柱子应用特殊标记 ---
    for j, bar in enumerate(bars):
        method_name = methods[j]
        
        if method_name == "Diffusion-DPO+Ours":
            bar.set_hatch('/')
            bar.set_edgecolor('white')
        
        height = bar.get_height()
        ax.annotate(f'{height:.2f}',
                    xy=(bar.get_x() + bar.get_width() / 2, height),
                    xytext=(0, 5), 
                    textcoords="offset points",
                    ha='center', va='bottom', 
                    fontsize=value_font_size, fontweight='bold', color='#333333')

    # 设置 y 轴保留一位小数
    ax.yaxis.set_major_formatter(mtick.FormatStrFormatter('%.1f'))
    ax.yaxis.set_major_locator(plt.MaxNLocator(nbins=4))
    
    # 标题加大并带箭头
    # ax.set_title(f"{metric}$\\uparrow$", fontsize=title_font_size, pad=10)
    ax.set_ylabel(f"{metric} $\\uparrow$", fontsize=title_font_size, labelpad=5)
    
    # 轴标签调整
    ax.tick_params(axis='x', labelsize=x_tick_font_size)
    ax.tick_params(axis='y', labelsize=y_tick_font_size)
    ax.set_xticklabels(formatted_methods, rotation=40, ha='right')
    
    # 优化显示范围：为顶部数字留出 30% 的额外空间
    min_v, max_v = min(values), max(values)
    y_range = max_v - min_v
    ax.set_ylim(max(0, min_v - y_range * 0.2), max_v + y_range * 0.3)
    
    # 细节美化
    # ax.yaxis.grid(True, linestyle='--', alpha=0.3)
    # ax.spines['top'].set_visible(False)
    # ax.spines['right'].set_visible(False)
    ax.set_axisbelow(True)

plt.tight_layout()
# plt.savefig('2026-ICML-SD-1.5-barchart-parti_prompts.svg', dpi=300, bbox_inches='tight')
plt.show()
```
