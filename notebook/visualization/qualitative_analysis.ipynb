```python
import random
import matplotlib.pyplot as plt
from PIL import Image

# Method dictionary and parameters (Separate w/o CFG and + CFG as different entries)
method_dict = {
    "sd-3-5-medium": [
        {"label": "(w/o CFG)", "ckpt": 0, "cfg_guidance": 1.0},
        {"label": "(+ CFG)", "ckpt": 0, "cfg_guidance": 4.5}
    ],
    "pickscore": [
        {"label": "(w/o CFG)", "ckpt": 150, "cfg_guidance": 1.0},
        {"label": "(+ CFG)", "ckpt": 150, "cfg_guidance": 4.5}
    ],
    "code": [
        {"label": "(w/o CFG)", "ckpt": 1, "cfg_guidance": 1.0},
        {"label": "(+ CFG)", "ckpt": 1, "cfg_guidance": 4.5}
    ]
    # "b_free": [
    #     # {"label": "(w/o CFG)", "ckpt": 1, "cfg_guidance": 1.0},
    #     {"label": "(+ CFG)", "ckpt": 2, "cfg_guidance": 4.5}
    # ]
}

dataset = "drawbench"
base_image_dir = f"/data_center/data2/dataset/chenwy/21164-data/diffusionnft/generate_images"
random.seed(42)  # For reproducibility
image_id_list = sorted(random.sample(range(0, 1000), 10))
print(f"selected image ids: {image_id_list}")

# Data structure to store the images and labels for each method
method_images = {f'{method}\n{item["label"]}': [] 
                 for method, params_list in method_dict.items() 
                 for item in params_list}

# Loop through methods, cfg_guidance, and image ids
for method, params_list in method_dict.items():
    for params in params_list:
        ckpt = params["ckpt"]
        cfg_guidance = params["cfg_guidance"]
        label = params["label"]
        
        for image_id in image_id_list:
            # Construct image path (assumes the image filenames are in the format "image_id.png")
            image_path = f"{base_image_dir}/sd3_cfg_{cfg_guidance}/{dataset}/{method}/ckpt-{ckpt}/images/{image_id:05d}.png"
            try:
                # Open the image file
                img = Image.open(image_path)
                method_images[f'{method}\n{label}'].append(img)
            except Exception as e:
                print(f"Error opening {image_path}: {e}")

# Calculate the total number of columns (method-cfg combinations)
num_columns = sum(len(params_list) for params_list in method_dict.values())
num_rows = len(image_id_list)

# Dynamically adjust figure size
fig_width = num_columns * 2  # 2 inches per column
fig_height = num_rows * 2  # 2 inches per row

# Create figure
plt.figure(figsize=(fig_width, fig_height), dpi=300)

# Create a flat list of method labels to maintain order
method_labels = []
for method, params_list in method_dict.items():
    for params in params_list:
        method_labels.append(f'{method}\n{params["label"]}')

# Plot images
for row_idx in range(num_rows):
    for col_idx, method_label in enumerate(method_labels):
        # Calculate subplot position (1-indexed)
        subplot_idx = row_idx * num_columns + col_idx + 1
        
        plt.subplot(num_rows, num_columns, subplot_idx)
        
        # Get the corresponding image
        if row_idx < len(method_images[method_label]):
            img = method_images[method_label][row_idx]
            plt.imshow(img)
        
        plt.axis('off')  # Hide axes
        
        # Add title only to the first row
        if row_idx == 0:
            plt.title(method_label, fontsize=10, pad=10)

plt.tight_layout()
plt.subplots_adjust(wspace=0.01, hspace=0.01)
plt.show()

# Optional: Save the figure
# plt.savefig('method_comparison.png', bbox_inches='tight', dpi=300)
```

```python
import random
import matplotlib.pyplot as plt
from PIL import Image

# Method dictionary and parameters (Separate w/o CFG and + CFG as different entries)
method_dict = {
    "sd-3-5-medium": [
        {"label": "(w/o CFG)", "ckpt": 0, "cfg_guidance": 1.0},
        {"label": "(+ CFG)", "ckpt": 0, "cfg_guidance": 4.5}
    ],
    "pickscore": [
        {"label": "(w/o CFG)", "ckpt": 150, "cfg_guidance": 1.0},
        {"label": "(+ CFG)", "ckpt": 150, "cfg_guidance": 4.5}
    ],
    "code": [
        {"label": "(w/o CFG)", "ckpt": 1, "cfg_guidance": 1.0},
        {"label": "(+ CFG)", "ckpt": 1, "cfg_guidance": 4.5}
    ]
}

dataset = "drawbench"
base_image_dir = f"/data_center/data2/dataset/chenwy/21164-data/diffusionnft/generate_images"
random.seed(42)  # For reproducibility
image_id_list = sorted(random.sample(range(0, 1000), 10))
print(f"selected image ids: {image_id_list}")

# Data structure to store the images and labels for each method
method_images = {f'{method}\n{item["label"]}': [] 
                 for method, params_list in method_dict.items() 
                 for item in params_list}

# Loop through methods, cfg_guidance, and image ids
for method, params_list in method_dict.items():
    for params in params_list:
        ckpt = params["ckpt"]
        cfg_guidance = params["cfg_guidance"]
        label = params["label"]
        
        for image_id in image_id_list:
            # Construct image path (assumes the image filenames are in the format "image_id.png")
            image_path = f"{base_image_dir}/sd3_cfg_{cfg_guidance}/{dataset}/{method}/ckpt-{ckpt}/images/{image_id:05d}.png"
            try:
                # Open the image file
                img = Image.open(image_path)
                method_images[f'{method}\n{label}'].append(img)
            except Exception as e:
                print(f"Error opening {image_path}: {e}")

# Calculate the total number of columns (method-cfg combinations)
num_columns = sum(len(params_list) for params_list in method_dict.values())
num_rows = len(image_id_list)

# Dynamically adjust figure size
fig_width = num_columns * 2  # 2 inches per column
fig_height = num_rows * 2  # 2 inches per row

# Create figure
plt.figure(figsize=(fig_width, fig_height), dpi=300)

# Reorganize method labels: first all (w/o CFG), then all (+ CFG)
method_labels_wo_cfg = []
method_labels_w_cfg = []

for method, params_list in method_dict.items():
    for params in params_list:
        label = f'{method}\n{params["label"]}'
        if "(w/o CFG)" in params["label"]:
            method_labels_wo_cfg.append(label)
        else:
            method_labels_w_cfg.append(label)

# Combine: all w/o CFG first, then all + CFG
method_labels = method_labels_wo_cfg + method_labels_w_cfg

# Plot images
for row_idx in range(num_rows):
    for col_idx, method_label in enumerate(method_labels):
        # Calculate subplot position (1-indexed)
        subplot_idx = row_idx * num_columns + col_idx + 1
        
        plt.subplot(num_rows, num_columns, subplot_idx)
        
        # Get the corresponding image
        if row_idx < len(method_images[method_label]):
            img = method_images[method_label][row_idx]
            plt.imshow(img)
        
        plt.axis('off')  # Hide axes
        
        # Add title only to the first row
        if row_idx == 0:
            plt.title(method_label, fontsize=10, pad=10)

plt.tight_layout()
plt.subplots_adjust(wspace=0.01, hspace=0.01)
plt.show()

# Optional: Save the figure
# plt.savefig('method_comparison.png', bbox_inches='tight', dpi=300)
```
