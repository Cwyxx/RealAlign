```python
### Multi-Step Compare, Single Image ####
import os
import math
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np

image_id = 5
method = "dinov2-default_config"
base_generated_image_dir = "/data_center/data2/dataset/chenwy/21164-data/flow-grpo/generate_images/sd3_textencoder_3_none_cfg_4.5/drawbench"
generated_image_dir = os.path.join(base_generated_image_dir, method)
ckpt_list = list(range(0, 321, 20))  # [0, 20, 40, ..., 240]

cols = 6
rows = math.ceil(len(ckpt_list) / cols)

fig, axes = plt.subplots(rows, cols, figsize=(3 * cols, 3 * rows), dpi=300)

axes = axes.flatten() if isinstance(axes, (list, np.ndarray)) else [axes]

for i, ckpt in enumerate(ckpt_list):
    img_path = os.path.join(generated_image_dir, f"ckpt-{ckpt}", "images", f"{image_id:05d}.png")

    img = Image.open(img_path)
    axes[i].imshow(img)
    axes[i].axis("off")
    if ckpt != 0:
        axes[i].set_title(f"ckpt {ckpt}")
    else:
        axes[i].set_title("SD-3-5-Medium")

for j in range(i + 1, rows * cols):
    axes[j].axis("off")

plt.tight_layout()
plt.show()
```
