```python
### Multi-Method Compare, Single Image ####
import os
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

base_image_dir = "/data_center/data2/dataset/chenwy/21164-data/diffusionnft/generate_images/sd3_textencoder_3_none_cfg_1.0/drawbench"

method_list = [ 
    "SD3.5M-DiffusionNFT-MultiReward", 
    # "sd3.5m-diffusionnft-multireward-next-code-win_0.5_3-lose_0.1_3-wo_normalize-wo_optprob",
    "sd3.5m-diffusionnft-multireward-next-code-win_0.5_3-lose_0.1_3-group_size_12-wo_normalize-wo_optprob",
    "sd3.5m-diffusionnft-multireward-next-code-win_0.4_3-lose_0.1_3-group_size_12-wo_normalize-wo_optprob",
    "sd3.5m-diffusionnft-multireward-next-code-win_0.3_3-lose_0.1_3-group_size_12-wo_normalize-wo_optprob",
]
ckpt_list = [0, 1, 1, 1]
image_id = 112
num_columns = 2
num_rows = (len(method_list) + 1) // num_columns  # 计算行数

fig, axes = plt.subplots(num_rows, num_columns, figsize=(10, 5 * num_rows),dpi=300)

if num_rows == 1:
    axes = [axes]
if num_columns == 1:
    axes = [axes]

for idx, (method, ckpt) in enumerate(zip(method_list, ckpt_list)):
    row = idx // num_columns
    col = idx % num_columns
    image_path = os.path.join(base_image_dir, method, f"ckpt-{ckpt}", "images", f"{image_id:05d}.png")
    img = mpimg.imread(image_path)
    axes[row][col].imshow(img)
    axes[row][col].axis('off')
    axes[row][col].set_title(method[len("SD3.5M-DiffusionNFT-MultiReward-next-code-"):-len("-wo_normalize-wo_optprob")])

plt.tight_layout()
plt.show()
```
