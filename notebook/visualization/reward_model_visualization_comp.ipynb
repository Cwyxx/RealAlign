```python
#### pickscore-analysis ####
import os
import pandas as pd
import matplotlib.pyplot as plt
from PIL import Image

score_func = [ "code", "dinov2", "pickscore", "deqa" ]
image_dir = "/data_center/data2/dataset/chenwy/21164-data/diffusionnft/generate_images/sd3_textencoder_3_none_cfg_1.0/pickscore-analysis/SD3.5M-DiffusionNFT-MultiReward/ckpt-0/images"
score_csv_file_path = f"/data_center/data2/dataset/chenwy/21164-data/diffusionnft/generate_images/sd3_textencoder_3_none_cfg_1.0/pickscore-analysis/SD3.5M-DiffusionNFT-MultiReward/ckpt-0/reward_score/{score_func}.csv"
df = pd.read_csv(score_csv_file_path)
prompt = "Fornasetti Style Picture of lost carthage  Westminster Abbey sunny roman villa burning, 18mm smoke filled room debris , gladiator ,floor mosaics  fire smoke, a photo,   roman , a digital rendering,  inside the roman colliseum,   plants overgrown outstanding detail ,room flooded, in front of a building, by PAUL ROBERTS"

df = df[df['prompt'] == prompt]
df = df.sort_values(by=score_func, ascending=False)
df_top = df.head(24)

rows = 4
cols = 6 
fig, axes = plt.subplots(rows, cols, figsize=(3 * cols, 3 * rows), dpi=300)
axes = axes.flatten()

for ax, (_, row) in zip(axes, df_top.iterrows()):
    image_path = os.path.join(image_dir, f"{row['image_name']}.png")
    if os.path.exists(image_path):
        img = Image.open(image_path)
        ax.imshow(img)
        ax.set_title(f"{score_func} score: {row[score_func]:.3f}", fontsize=9)
    else:
        ax.text(0.5, 0.5, "Missing", ha='center', va='center', fontsize=10, color='red')
    ax.axis('off')

for i in range(len(df_top), len(axes)):
    axes[i].axis('off')

plt.tight_layout()
plt.show()
```

```python
#### pickscore-analysis-multi_model ####
import os
import pandas as pd
import matplotlib.pyplot as plt
from PIL import Image
from functools import reduce

score_func = ["dinov2", "code", "aesthetic", "aesthetic_v2_5", "clip_iqa", "imagereward"]
base_dir = "/data_center/data2/dataset/chenwy/21164-data/diffusionnft/generate_images/sd3_textencoder_3_none_cfg_1.0/geneval-analysis/SD3.5M-DiffusionNFT-MultiReward/ckpt-0"
image_dir = os.path.join(base_dir, "images")
score_dir = os.path.join    (base_dir, "reward_score")

dfs = []
for func in score_func:
    path = os.path.join(score_dir, f"{func}.csv")
    df = pd.read_csv(path)[["prompt", "image_name", func]]
    dfs.append(df)
merged_df = reduce(lambda left, right: pd.merge(left, right, on=['prompt', 'image_name'], how='outer'), dfs)
df = merged_df

prompt = "a photo of an umbrella and an airplane"
df = df[df["prompt"] == prompt]
df = df.sort_values(by="dinov2", ascending=False)
df_top = df.head(24)

rows, cols = 4, 6
fig, axes = plt.subplots(rows, cols, figsize=(3 * cols, 3 * rows), dpi=300)
axes = axes.flatten()

for ax, (_, row) in zip(axes, df_top.iterrows()):
    image_path = os.path.join(image_dir, f"{row['image_name']}.png")
    if os.path.exists(image_path):
        img = Image.open(image_path)
        ax.imshow(img)
        score_text = "\n".join([f"{func}: {row[func]:.3f}" for func in score_func])
        ax.text(
            0.02, 0.98, score_text,
            transform=ax.transAxes,
            fontsize=7, color="white", va="top", ha="left",
            bbox=dict(facecolor="black", alpha=0.5, pad=2)
        )
    else:
        ax.text(0.5, 0.5, "Missing", ha="center", va="center", fontsize=10, color="red")
    ax.axis("off")

for i in range(len(df_top), len(axes)):
    axes[i].axis("off")

plt.tight_layout()
plt.show()
```
