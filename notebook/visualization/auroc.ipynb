```python
import pandas as pd
from sklearn.metrics import roc_curve, auc, accuracy_score
import matplotlib.pyplot as plt

# 读取数据
base_csv_path = "/data3/chenweiyan/notebook/fine-tune-diffusion/spo_gitee/tmp/SD_3_5_medium/detect_score/code.csv"
df = pd.read_csv(base_csv_path)

# 提取真实标签和预测值
y_true = df["image_type"].tolist()
y_pred = df["model_output"].tolist()

accuracy = accuracy_score(y_true, [1 if prob > 0.5 else 0 for prob in y_pred])
print(f"Accuracy: {accuracy}")

fpr, tpr, thresholds = roc_curve(y_true, y_pred)

auroc = auc(fpr, tpr)
print(f"AUROC: {auroc}")

# 绘制 ROC 曲线
plt.figure(figsize=(8, 6), dpi=300)
plt.plot(fpr, tpr, color='blue', label=f'ROC curve (AUROC = {auroc:.2f})')
plt.plot([0, 1], [0, 1], color='gray', linestyle='--')  # 对角线
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc='lower right')
plt.show()
```
