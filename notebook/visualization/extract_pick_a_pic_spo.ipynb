```python
import os
import json

source_json_path="/data_center/data2/dataset/chenwy/21164-data/dpo_dataset/pick_a_pic_v1/pick_a_pic_validation_prompt_500.json"
with open(source_json_path, 'r') as file:
    data = json.load(file)

base_prompt_dir="/data3/chenweiyan/notebook/fine-tune-diffusion/spo_gitee/DiffusionNFT/dataset/pick_a_pic_spo"
os.makedirs(base_prompt_dir, exist_ok=True)
prompt_path = os.path.join(base_prompt_dir, 'test.txt')
with open(prompt_path, 'w') as output_file:
    for item in data:
        caption = item.get('caption', '')  # 提取每个对象中的caption
        output_file.write(caption + '\n')  # 将caption写入文本文件，每个caption占一行
```
