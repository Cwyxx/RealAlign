```python
import os
source_csv_file_path = "/data3/chenweiyan/notebook/fine-tune-diffusion/OneIG-Benchmark/OneIG-Bench.csv"
target_txt_file_path = "/data3/chenweiyan/notebook/fine-tune-diffusion/spo_gitee/DiffusionNFT/dataset/OneIG-Bench-Portrait/test.txt"
os.makedirs(os.path.dirname(target_txt_file_path), exist_ok=True)

import pandas as pd

# 读取 CSV 文件
df = pd.read_csv(source_csv_file_path)

# 筛选 category 为 "Anime_Stylization" 的行
anime_df = df[df['category'] == 'Portrait']

# 提取 prompt_en 列
prompts = anime_df['prompt_en'].tolist()

# 写入到 txt 文件，每行一个 prompt
with open(target_txt_file_path, 'w', encoding='utf-8') as f:
    for prompt in prompts:
        f.write(prompt + '\n')

print(f"已提取 {len(prompts)} 条 Portrait 的 prompt_en 到 {target_txt_file_path}")
```
